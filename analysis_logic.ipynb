```python
import requests
from bs4 import BeautifulSoup
from binance.client import Client
import pandas as pd
from datetime import datetime, timedelta
```

```python
api_key = ''
api_secret = ''

def main():
    client = Client(api_key, api_secret)

    res = client.get_exchange_info()
    print(client.response.headers)

if __name__ == "__main__":
    main()
```

```python
client = Client(api_key, api_secret)
data_700 = client.get_historical_klines("BTCUSDT", Client.KLINE_INTERVAL_4HOUR, "600 day ago UTC")
```

```python
columns = ['Open time','Open','High','Low','Close','Volume','Close time','Quote asset volume','Number of trades','Taker buy base asset volume','Taker buy quote asset volume','Ignore']
goo = pd.DataFrame(data_700,columns=columns)
```

```python
goo['Open time'] = goo['Open time'].apply(lambda x : datetime.fromtimestamp(x/1000))
goo['Close time'] = goo['Close time'].apply(lambda x : datetime.fromtimestamp(x/1000))
```

```python
goo.head()
```

```python
goo.info()
goo['Close'] = goo['Close'].astype(float)
```

```python
goo['change_7'] = goo['Close'].pct_change(7)
goo['buy'] = goo['Close'].shift(-1)
goo['day'] = goo['Open time'].dt.strftime('%d/%m/%y')
goo.head(10)
```

```python
foo_2 = goo[goo.day == '04/12/21']

import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=foo_2['Open time'],
                open=foo_2['Open'],
                high=foo_2['High'],
                low=foo_2['Low'],
                close=foo_2['Close'])])

fig.show()
```

```python
df_d = goo[goo['change_7'] < -0.07]
df_d.groupby('day',as_index=False)['Open time'].count().sort_values('day')
```

```python
df_d = goo[goo['change_7'] < -0.06]
df_d.groupby('day',as_index=False)['Open time'].count().sort_values('day')
```
